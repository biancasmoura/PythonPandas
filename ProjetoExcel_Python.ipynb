**Trabalhando com Planilhas do Excel**

```python
#importando a biblioteca

import pandas as pd
```

```python
#Leitura arquivos
df1 = pd.read_excel("/content/Aracaju.xlsx")
df2 = pd.read_excel("/content/Fortaleza.xlsx")
df3 = pd.read_excel("/content/Natal.xlsx")
df4 = pd.read_excel("/content/Recife.xlsx")
df5 = pd.read_excel("/content/Salvador.xlsx")
```

```python
#ver os df`s

df1.head(1)
df2.head(1)
df3.head(1)
df4.head(1)
df5.head(1)
```

```python
#Juntar todos os arquivos

df = pd.concat([df1, df2, df3,df4, df5])
```

```python
#exibindo as 5 primeiras linhas
df.head()
```

```python
#ver as 5 últimas linhas

df.tail()
```

```python
#pegar exemplos da planilha
df.sample(5)
```

```python
#verificar os tipos

df.dtypes
```

```python
#Alterar o tipo de dado de cada coluna

df["lojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

**Tratando valores faltantes**

```python
#Consultando linhas faltantes

df.isnull().sum()
```

```python
#distribuir os valores nulos pela média

df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#Apagar valores nulos
df.dropna(inplace=True)
```

**Criando colunas**

```python
#Criando uma coluna de receita

df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#Retorna as 3 linhas com maiores receitas
df.nlargest(3, "Receita")
```

```python
#Retorna as 3 linhas com menores receitas
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordenando conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

**Trabalhando com Datas**

```python
#Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificando o tipo de dado em cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data 
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por ano

df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["mes_venda"],df["dia_venda"] = (df["Data"].dt.month,df["Data"].dt.day )
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga

df["Data"].min()
```

```python
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] -df["Data"].min()
df.sample(5)
```

```python
#Criando coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19.sample(20)
```
